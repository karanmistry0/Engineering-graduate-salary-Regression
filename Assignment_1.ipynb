```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
sns.histplot(data['Salary'], bins=30, kde=True, color='blue')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='CollegeTier', y='Salary', data=data, palette='coolwarm')
plt.title("Salary Distribution by College Tier")
plt.xlabel("College Tier")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='collegeGPA', y='Salary', data=data, alpha=0.6)
plt.title("College GPA vs. Salary")
plt.xlabel("College GPA")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Specialization', y='Salary', data=data, estimator=lambda x: x.mean(), palette='viridis')
plt.xticks(rotation=90)
plt.title("Average Salary by Specialization")
plt.xlabel("Specialization")
plt.ylabel("Average Salary")
plt.show()
```

```python
data = data.dropna(axis=1)
data.columns
```

```python
data = data.loc[:, ~data.columns.duplicated()]
data.columns
```

```python
data.fillna(data.mean(numeric_only=True), inplace=True)
```
